# Pregunta 2 - Funciones

Cree una función que reciba de entrada la edad de 3 personas, y que retorne el promedio de las 2 personas de mayor edad.

```python
def media_edad(a,b,c):
   lista_edades = [a,b,c]
   lista_edades.sort(reverse=True)
   print(lista_edades)
   mayor1=lista_edades[0]
   mayor2=lista_edades[1]
   media=(mayor1+mayor2)/2
   return media

```

```python
media_edad(34,8,67)
```
